```python
import pandas as pd
```

```python
df=pd.read_csv("Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
#using this function we know that we should deal with TotalCharges column
```

```python
non_numeric = pd.to_numeric(df['TotalCharges'], errors='coerce').isna()
unique_non_numerics = df.loc[non_numeric, 'TotalCharges'].unique()
unique_non_numerics
```

```python
# Check rows where TotalCharges is just empty spaces
empty_charges = df[pd.to_numeric(df['TotalCharges'],errors="coerce" ).isnull()]
print(f"Number of rows with empty spaces: {len(empty_charges)}")
print(empty_charges[['tenure', 'MonthlyCharges', 'TotalCharges']])
#from this code we can see why the totalCharges = " " because the cus did not complete their first month in subscription
#for that we will replace that " " which will be NaN with zero
```

```python
#by this one we convert it to NaN
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors="coerce")
#this one errors ='raise' check if every thing worked good
check_clean=pd.to_numeric(df['TotalCharges'],errors='raise')
```

```python
#to check that all the nan is the  " " not anything else
df['TotalCharges'].isna().sum()
```

```python
df['TotalCharges']=df['TotalCharges'].fillna(0.0)
```
